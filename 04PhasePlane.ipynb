# Drawing Phase Plane for System of Ordinary Differential Equations

```python
import numpy as np
from matplotlib import pyplot as plt
import mpl_config
```

## Competitive predator model

```python
a = 1
b = 1
m = 1
n = 1
```

```python
x = np.linspace(0, 3)
y = np.linspace(0, 3)

u = a * x.reshape(1, -1) - b * x.reshape(1, -1) * y.reshape(-1, 1)
v = m * y.reshape(-1, 1) - n * x.reshape(1, -1) * y.reshape(-1, 1)
```

```python
plt.streamplot(x, y, u, v, broken_streamlines=False, density=0.5)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

## Predator and pray model

```python
a = 1
b = 1
m = 1
n = 1
```

```python
x = np.linspace(0, 3)
y = np.linspace(0, 3)

u = a * x.reshape(1, -1) - b * x.reshape(1, -1) * y.reshape(-1, 1)
v = -m * y.reshape(-1, 1) + n * x.reshape(1, -1) * y.reshape(-1, 1)
```

```python
plt.streamplot(x, y, u, v, broken_streamlines=False, density=0.5)
plt.xlabel("$x$ (Krill)")
plt.ylabel("$y$ (Whale)")
plt.show()
```
