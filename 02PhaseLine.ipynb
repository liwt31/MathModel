# Drawing Phase Line for Ordinary Differential Equation
We are going to use `streamplot` from `matplotlib`. Learn to use Google and the Documenetation.

```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use("mm.mplstyle")
```

## A simple example
$$\frac{dy}{dx} = (y+1)(y-2)$$

```python
x = np.linspace(-5, 5)
y = np.linspace(-5, 5)
```

```python
# Note that y for the row and x for the col.
# The same as coordination axes
u = np.ones((len(y), len(x)))
v = ((y + 1 ) * (y - 2)).reshape(-1, 1) * np.ones(len(x)).reshape(1, -1)
```

```python
plt.streamplot(x, y, u, v, broken_streamlines=False, density=0.5)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

## Logistic population growth model

```python
# the model parameters
M = 665
r = 0.53 / M
```

```python
t = np.linspace(0, 18)
p = np.linspace(0, 1.3 * M)

u = np.ones((len(p), len(t)))
v = (r * p * (M - p)).reshape(-1, 1) * np.ones(len(t)).reshape(1, -1)
```

```python
plt.streamplot(t, p, u, v, broken_streamlines=False, density=0.5)
plt.xlabel("$t$")
plt.ylabel("$p$")
plt.show()
```
