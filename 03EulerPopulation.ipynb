# Population Growth with Euler's Method

```python
import numpy as np
from matplotlib import pyplot as plt
import mpl_config
```

```python
# the model parameters
M = 665
r = 0.53 / M
P0 = 9.6
```

```python
dt = 0.1
```

```python
t = 0
p = P0
t_list = []
p_list = []
while t < 18:
    p += r * p * (M - p) * dt
    t += dt
    t_list.append(t)
    p_list.append(p)
```

```python
plt.plot(t_list, p_list, label="simulated")
time = np.array(t_list)
plt.plot(time, M * P0 / (P0 + (M-P0) * np.exp(-r*M*time)), label="analytical")
plt.xlabel("Time (h)")
plt.ylabel("Yeat biomass")
plt.legend()
plt.show()
```
