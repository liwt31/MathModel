# Predator and Pray Model with Euler's Method

```python
import numpy as np
from matplotlib import pyplot as plt
import mpl_config
```

```python
# the model parameters
a = 3
b = 1
m = 2
n = 1
```

```python
dt = 0.1
```

```python
t = 0
# krill
x0 = 1
# whale
y0 = 2
t_list = [t]
x_list = [x0]
y_list = [y0]
while t < 3:
    x = x0 + (a * x0 - b * x0 * y0) * dt
    y = y0 + (-m * y0 + n * x0 * y0) * dt
    x0 = x
    y0 = y
    t += dt
    x_list.append(x)
    y_list.append(y)
    t_list.append(t)
```

```python
plt.scatter(x_list, y_list)
plt.xlabel("Krill")
plt.ylabel("Whale")
plt.show()
```

```python
plt.plot(t_list, x_list, label="Krill")
plt.plot(t_list, y_list, label="Whale")
plt.xlabel("Time")
plt.ylabel("Population")
plt.legend()
plt.show()
```
