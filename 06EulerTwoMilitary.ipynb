# Predator and Pray Model with Euler's Method

```python
import numpy as np
from matplotlib import pyplot as plt
import mpl_config
```

```python
# the model parameters
a = 5
b = 1
```

```python
dt = 0.1
```

```python
t = 0
# krill
x0 = 2
# whale
y0 = 1
t_list = [t]
x_list = [x0]
y_list = [y0]
while t < 3 and x0 > 0 and y0 > 0:
    x = x0 + (-a * y0) * dt
    y = y0 + (-b * x0) * dt
    x0 = x
    y0 = y
    t += dt
    x_list.append(x)
    y_list.append(y)
    t_list.append(t)
```

```python
plt.scatter(x_list, y_list, c=t_list)
plt.xlabel("Force $x$")
plt.ylabel("Force $y$")
plt.colorbar()
plt.show()
```

Problems
- Why add the condition of $x > 0$ and $y > 0$ for the loop? (what if they are not there?)
- What does the color mean? (what if there is no color?)

```python
C = a * y_list[0]**2 - b * x_list[0]**2
if C > 0:
    remaining_force = np.sqrt(C / a)
else:
    remaining_force = np.sqrt(-C / b)
plt.hlines(remaining_force, t_list[0], t_list[-1], linestyle="--", color="black")

plt.plot(t_list, x_list, label="Force $x$")
plt.plot(t_list, y_list, label="Force $y$")

plt.xlabel("Time")
plt.ylabel("Strength of force")
plt.legend()
plt.show()
```
