# Elapsed time of a tape recorder
Popynomial fit

## The data

```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use("mm.mplstyle")
```

```python
c = np.array([100, 200, 300, 400, 500, 600, 700, 800])
c = c / 100
c
```

```python
t = np.array([205, 430, 677, 945, 1233, 1542, 1872, 2224, ])
```

## Ploting the data

```python
plt.plot(c, t, marker="o")
plt.xlabel("Counter")
plt.ylabel("Elapsed Time")
plt.show()
```

## Fit the data with high order polynomial

```python
coeffs = np.polyfit(c, t, deg=7)
coeffs
```

```python
x = np.linspace(0, 10)
y = np.zeros(len(x))
for i in range(8):
    y += x ** (7-i) * coeffs[i]

plt.plot(x, y, label="Fit")

plt.scatter(c, t, marker="o", label="Data")
plt.xlabel("Counter")
plt.ylabel("Elapsed Time")
plt.legend()
plt.show()
```
